```python
import pandas as pd
```

```python
# botometer-feedback-2019
# Description: Botometer feedback accounts manually labeled by K.C. Yang. Labels and user objects.

botometer_feedback_2019 = pd.read_json('botometer-feedback-2019/botometer-feedback-2019_tweets.json', orient='records')
botometer_2019_df = pd.DataFrame(list(botometer_feedback_2019['user']))
botometer_2019_df['collected_at'] = botometer_feedback_2019['created_at']

# data labels
botometer_2019_labels = pd.read_csv('botometer-feedback-2019/botometer-feedback-2019.tsv', sep='\t', header=None)
botometer_2019_labels.columns = ['id', 'label']

# merge in labels on id
botometer_2019_df = botometer_2019_df.merge(botometer_2019_labels, how='left')

botometer_2019_df
```

```python
# botwiki-2019
# Description: Self-identified bots from https://botwiki.org. Labels and user objects.
botwiki_2019_tweets = pd.read_json('botwiki-2019/botwiki-2019_tweets.json')
botwiki_2019_df = pd.DataFrame(list(botwiki_2019_tweets['user']))
botwiki_2019_df['collected_at'] = botwiki_2019_tweets['created_at']
botwiki_2019_df

# data labels
botwiki_2019_labels = pd.read_csv('botwiki-2019/botwiki-2019.tsv', sep='\t', header=None)
botwiki_2019_labels.columns = ['id', 'label']

# merge in labels on id
botwiki_2019_df = botwiki_2019_df.merge(botwiki_2019_labels, how='left')

botwiki_2019_df
```

```python
# celebrity-2019
# Description: Celebrity accounts collected as authentic users. Labels and user objects.
celebrity_2019_tweets = pd.read_json('celebrity-2019/celebrity-2019_tweets.json', orient='records')
celebrity_2019_df = pd.DataFrame(list(celebrity_2019_tweets['user']))
celebrity_2019_df['collected_at'] = celebrity_2019_tweets['created_at']

# data labels
celebrity_2019_labels = pd.read_csv('celebrity-2019/celebrity-2019.tsv', sep='\t', header=None)
celebrity_2019_labels.columns = ['id', 'label']

# merge in labels on id
celebrity_2019_df = celebrity_2019_df.merge(celebrity_2019_labels, how='left')

# they're all human in this df but some show up as NaN after the merge
celebrity_2019_df.fillna('human', inplace=True)

celebrity_2019_df
```

```python
# cresci-2015
# Description: A dataset of (i) genuine and (ii) fake Twitter accounts, manually annotated. Released in CSV format.
# https://arxiv.org/pdf/1509.04098.pdf

# human
cresci_2015_E13 = pd.read_csv('cresci-2015/E13/users.csv')
cresci_2015_E13['label'] = 'human'

# bots
cresci_2015_FSF = pd.read_csv('cresci-2015/FSF/users.csv')
cresci_2015_FSF['label'] = 'bot'

# bots
cresci_2015_INT = pd.read_csv('cresci-2015/INT/users.csv')
cresci_2015_INT['label'] = 'bot'

# human
cresci_2015_TFP = pd.read_csv('cresci-2015/TFP/users.csv')
cresci_2015_TFP['label'] = 'human'

# bots
cresci_2015_TWT = pd.read_csv('cresci-2015/TWT/users.csv')
cresci_2015_TWT['label'] = 'bot'

# combine into one df
cresci_2015_df = pd.concat([cresci_2015_E13, cresci_2015_FSF, cresci_2015_INT, cresci_2015_TFP, cresci_2015_TWT])
cresci_2015_df.rename(columns={'updated':'collected_at'}, inplace=True)
cresci_2015_df
```

```python
# cresci-2017
# Description: A dataset of (i) genuine, (ii) traditional, and (iii) social spambot Twitter accounts, annotated by CrowdFlower contributors. Released in CSV format.
    
social_spambots_1 = pd.read_csv('cresci-2017/datasets_full/social_spambots_1/users.csv')
social_spambots_2 = pd.read_csv('cresci-2017/datasets_full/social_spambots_2/users.csv')
social_spambots_3 = pd.read_csv('cresci-2017/datasets_full/social_spambots_3/users.csv')
traditional_spambots_1 = pd.read_csv('cresci-2017/datasets_full/traditional_spambots_1/users.csv')
traditional_spambots_2 = pd.read_csv('cresci-2017/datasets_full/traditional_spambots_2/users.csv')
traditional_spambots_3 = pd.read_csv('cresci-2017/datasets_full/traditional_spambots_3/users.csv')
traditional_spambots_4 = pd.read_csv('cresci-2017/datasets_full/traditional_spambots_4/users.csv')

#traditional_spambots_1 has errored created_at data, using 'timestamp' instead
traditional_spambots_1['created_at'] = traditional_spambots_1['timestamp']

# label genuine dataset
cresci_2017_genuine = pd.read_csv('cresci-2017/datasets_full/genuine_accounts/users.csv')
cresci_2017_genuine['label'] = 'human'

# combine and label bot datasets
cresci_2017_bots = pd.concat([social_spambots_1, social_spambots_2, social_spambots_3, traditional_spambots_1, traditional_spambots_2, traditional_spambots_3, traditional_spambots_4])
cresci_2017_bots['label'] = 'bot'

# combine into one dataset
cresci_2017_df = pd.concat([cresci_2017_genuine, cresci_2017_bots])
# using 'updated' instead of 'crawled_at'
cresci_2017_df.rename(columns={'updated':'collected_at'}, inplace=True)
cresci_2017_df
```

```python
# caverlee-2011
# Description: This social honeypot dataset collected from December 30, 2009 to August 2, 2010 on Twitter. 
# The dataset contains 22,223 content polluters, their number of followings over time 2,353,473 tweets, 
# and 19,276 legitimate users, their number of followings over time and 3,259,693 tweets.
# file:///Users/Kelso/Downloads/Seven_Months_with_the_Devils_A_Long-Term_Study_of_.pdf
# Note: missing most metadata features, likely not usable
honeypot_2011_bots_df = pd.read_csv('social_honeypot_icwsm_2011/content_polluters.txt', sep='\t', header=None)
honeypot_2011_bots_df.columns=['UserID', 'created_at', 'collected_at', 'following_count', 'followers_count', 'tweets_count', 'screen_name_length', 'profile_desc_length']

honeypot_2011_legitimate_df = pd.read_csv('social_honeypot_icwsm_2011/legitimate_users.txt', sep='\t', header=None)
honeypot_2011_legitimate_df.columns=['UserID', 'created_at', 'collected_at', 'following_count', 'followers_count', 'tweets_count', 'screen_name_length', 'profile_desc_length']

# labels
honeypot_2011_bots_df['label'] = 'bot'
honeypot_2011_legitimate_df['label'] = 'human'

honeypot_2011_df = pd.concat([honeypot_2011_bots_df, honeypot_2011_legitimate_df])

honeypot_2011_df.rename(columns={'UserID':'id', 'following_count':'friends_count', 'tweets_count':'statuses_count'}, inplace=True)

honeypot_2011_df
# follower counts over time
# pd.read_csv('social_honeypot_icwsm_2011/content_polluters_followings.txt', sep='\t', header=None)
# pd.read_csv('social_honeypot_icwsm_2011/legitimate_users_followings.txt', sep='\t', header=None)
```
